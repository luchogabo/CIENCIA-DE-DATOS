```python
# title@ **e. Coeficiente de variación**
```



La **fórmula del coeficiente de variación $CV$** es la siguiente:

$$
CV = \frac{\sigma}{\mu} \times 100
$$

Donde:
- $ \sigma $ es la desviación estándar de los datos.
- $ \mu $ es la media de los datos.
- El resultado se multiplica por 100 para expresarlo como un porcentaje.

### **Importancia del coeficiente de variación:**
1. **Medida de dispersión relativa:** El CV permite comparar la variabilidad de dos o más conjuntos de datos, incluso si tienen diferentes unidades o escalas. A diferencia de la desviación estándar, que da una medida absoluta de la dispersión, el CV normaliza esta variabilidad respecto a la media.
   
2. **Facilidad de interpretación:** Al expresarse como un porcentaje, es fácil de entender cuánta variabilidad relativa existe respecto a la media.

3. **Comparación de diferentes distribuciones:** El CV es útil cuando queremos comparar la consistencia o variabilidad de diferentes conjuntos de datos, como inversiones con diferentes medias de rendimiento o experimentos científicos con resultados en diferentes unidades.

```python
# @title **e. Coeficiente de variación**



# Calculamos el coeficiente de variación para cada variable

a = (df["DEFINITIVA"].std() / df["DEFINITIVA"].mean()) * 100
b = (df["ASISTENCIA"].std() / df["ASISTENCIA"].mean()) * 100
c = (df["EDAD"].std() / df["EDAD"].mean()) * 100
d = (df["PESO"].std() / df["PESO"].mean()) * 100
e = (df["ESTATURA"].std() / df["ESTATURA"].mean()) * 100


print('los coeficientes de variación de cada variable son')
print("DEFINITIVA", round(a, 2),'%')
print("ASISTENCIA", round(b, 2),'%')
print("EDAD", round(c, 2),'%')
print("PESO", round(d, 2), '%')
print("ESTATURA", round(d, 2),'%')
```
